```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create data

x = np.linspace(0, 10, 50)
y = np.cos(x)
```

## First Plot

```python
plt.plot(x, y)
```

```python
# set new figure size
fig = plt.figure(figsize=(6,3)) # inches
plt.plot(x, y, "--o", label=r"$y$") # linestyles: "-", "--", "-.", ":" 
plt.plot(x,y*2, ":+", label=r"$2\times \alpha$")

plt.legend(loc=0)
```

- axes labels
- legend
- title
- arrow end axis
- subplots

```python
# set new figure size
fig = plt.figure(figsize=(6,3)) # inches
# plt.plot(x, y, "--o", markerfacecolor="#70738c", markeredgecolor="yellow", label=r"$y$") # linestyles: "-", "--", "-.", ":" 
plt.plot(x, y, "--o", mfc="#70738c", mec="yellow", ms=10, mew=4, label=r"$y$") # linestyles: "-", "--", "-.", ":" 
plt.plot(x,y*2, "-+", color="black", mec="orange", ms=10, label=r"$2\times \alpha$")

plt.legend(loc=0)
```

```python
# Set axis limits
fig = plt.figure(figsize=(6,3)) # inches
plt.plot(x,y)
plt.xlim((2.5, 7.5))
plt.ylim((0, 1))

# axis labels
plt.xlabel("x", fontsize=20)
plt.ylabel(r"$\cos(x)$", color="r")

plt.savefig("myplot.png", dpi=300)
plt.savefig("myplot.pdf")
plt.close()
```
